```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
# Load the cascade
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
# Read the input image
image = cv.imread('/content/drive/MyDrive/face/bang5.jpg')
# Convert into grayscale
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
# Detect faces
faces = face_cascade.detectMultiScale(gray,scaleFactor= 1.1,minNeighbors= 8)
# Process each detected face
for (x, y, w, h) in faces:
    # Crop the face from the original image
    face = image[y:y+h, x:x+w]
# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv.rectangle(image, (x, y), (x+w, y+h), (255, 0, 255), 4)
# Display all faces
cv2_imshow(image)
for (x, y, w, h) in faces:
    face = image[y:y+h, x:x+w]
    plt.imshow(cv.cvtColor(face, cv.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
# Display the output
num_faces = len(faces)
print(f"จำนวนคนในภาพ: {num_faces} คน")




```
